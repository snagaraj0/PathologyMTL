```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from PIL import ImageOps
import os
import numpy as np
import skimage.io as io
```

```python
Image.MAX_IMAGE_PIXELS = None
mask = np.asarray(Image.open("../Downloads/7ad575414ac3516485d5ed13602aa8b8_mask.tiff"))
```

```python
# Plot each channel separately
for channel in range(mask.shape[-1]):
    plt.subplot(1, mask.shape[-1], channel + 1)
    plt.imshow(mask[:, :, channel], cmap='viridis')
    plt.axis('off')
    plt.title('Channel {}'.format(channel + 1))

# Combine channels into a single image
combined_mask = mask.max(axis=-1)
plt.figure()
plt.imshow(combined_mask, cmap='viridis')
plt.axis('off')
plt.title('Combined Mask')

plt.show()
```
